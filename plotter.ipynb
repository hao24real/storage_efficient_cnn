```python
import matplotlib.pyplot as plt

# Data
models = ['Base Model', 'Model V1', 'Model V2', 'Model V3', 'Model V4', 'Model V5', 'Model V6', 'Model Alpha']
validation_accuracy = [73.09, 72.78, 73.71, 70.39, 70.10, 68.03, 80.67, 81.04]
inference_time = [0.408, 0.755, 0.414, 0.781, 1.192, 1.029, 0.848, 1.400]

# Plotting
plt.figure(figsize=(10, 6))
for i, model in enumerate(models):
    plt.scatter(validation_accuracy[i], inference_time[i], label=model)

plt.xlabel('Validation Accuracy (%)')
plt.ylabel('Inference Time (ms)')
plt.title('Model Comparison: Validation Accuracy vs Inference Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Further adjusting text placement to avoid overlap, especially for Base Model and Model V2
plt.figure(figsize=(10, 6))
for i, model in enumerate(models):
    # Adjusting offset for specific models to avoid overlap
    offset_x = 0.2
    offset_y = 0.0
    if model == "Base Model":
        offset_y = -0.05  # Move text down for Base Model
    elif model == "Model V2":
        offset_y = 0.05  # Move text up for Model V2

    plt.scatter(validation_accuracy[i], inference_time[i])
    plt.text(validation_accuracy[i] + offset_x, inference_time[i] + offset_y, model, fontsize=9, ha='left')

plt.xlabel('Validation Accuracy (%)')
plt.ylabel('Inference Time (ms)')
plt.title('Model Comparison: Validation Accuracy vs Inference Time')
plt.grid(True)
plt.show()
```
